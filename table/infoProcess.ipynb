```python
from paddleocr import PaddleOCR, draw_ocr
ocr = PaddleOCR(use_angle_cls=True, lang="ch")
result = ocr.ocr("data/北医三院jpg/20230117-9_0.jpg", cls=True)
cordinates = []
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        cordinates.append(line)
print(cordinates)
```
